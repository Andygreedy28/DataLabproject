```python
start_time = '2022-02-24T00:00:00Z' #begining of Russia's "special military operations in Ukraine"
end_time = '2022-04-09T00:00:00Z'
columns=["Time","User_ID","Tweet", "like_count", "reply_count","retweet_count","language" ]
data=[]
for tweet in tweepy.Paginator(client.get_users_tweets, id=roblee_user_id,exclude="retweets",start_time=start_time,
                 end_time=end_time, tweet_fields=['context_annotations','author_id','created_at','geo','public_metrics','lang'],user_fields=['profile_image_url'], max_results=100).flatten(limit=3200):
   data.append([tweet.created_at,tweet.author_id,tweet.text,tweet.public_metrics["like_count"],tweet.public_metrics["reply_count"],tweet.public_metrics["retweet_count"],tweet.lang])
  
roblee_df=pd.DataFrame(data,columns=columns) #remark: data was restricted to latest 3200tweets only
roblee_df=roblee_df.sort_values(by='Time')
roblee_df=roblee_df.set_index("Time")

roblee_df.to_csv("roblee_tweet.csv") #save as csv file


roblee_df.head(10)
    

```
